```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
iris = datasets.load_iris()
X = iris.data  
y = iris.target 
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
k = 3  
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X_scaled)

labels = kmeans.labels_
```

```python
df = pd.DataFrame(X_scaled, columns=iris.feature_names)
df['Cluster'] = labels

plt.figure(figsize=(10, 6))

plt.subplot(2,2,1)
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=df['Cluster'], cmap='viridis', marker='o')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200)
plt.title('K-means Clustering of Iris Dataset')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.grid()

plt.subplot(2,2,2)
plt.scatter(df.iloc[:, 1], df.iloc[:, 2], c=df['Cluster'], cmap='viridis', marker='o')
plt.scatter(kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 2], c='red', marker='X', s=200)
plt.title('K-means Clustering of Iris Dataset')
plt.xlabel(iris.feature_names[1])
plt.ylabel(iris.feature_names[2])
plt.grid()

plt.subplot(2,2,3)
plt.scatter(df.iloc[:, 2], df.iloc[:, 3], c=df['Cluster'], cmap='viridis', marker='o')
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:, 3], c='red', marker='X', s=200)
plt.title('K-means Clustering of Iris Dataset')
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.grid()

plt.subplot(2,2,4)
plt.scatter(df.iloc[:, 1], df.iloc[:, 3], c=df['Cluster'], cmap='viridis', marker='o')
plt.scatter(kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 3], c='red', marker='X', s=200)
plt.title('K-means Clustering of Iris Dataset')
plt.xlabel(iris.feature_names[1])
plt.ylabel(iris.feature_names[3])
plt.grid()

plt.tight_layout()
plt.show()
```
